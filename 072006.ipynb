```python
'''
x = x + 1, x > 0
终止循环 x <= 0
输出 "while end"
'''

running = True
while running:
    x = int(input('please enter the x: '))
    if x > 0:
        print(x + 1)
    else:
        running = False
else:
    print('while end')
```

```python
'''
输入一堆字符串
每行一个
输出每个字符串的长度
如果字符串长度小于等于3 则输出太短了
以quit截止
while boolean_expression:
    code
'''

while True:
    x = input('please enter a string: ')
    if x == 'quit':
        break
    if len(x) <= 3:
        print('too short')
        continue
    print(len(x))
```

```python
'''
输入一串数字，每行一个数字
如果是偶数，则输出该数字
否则不输出
以-1结束输入

break and continue
'''

while True:
    x = input()
    x = int(x)
    # x = int(input())
    if x == -1:
        break
    if x % 2 == 0:
        print(x)
```

```python
"""
输入一个正整数判断它是不是质数

循环 (1,t)
如果有能整除的数 就不是质数
"""

t = input('please input a number: ')
t = int(t)
if t == 1:
    print(False)
else:
    for i in range(2, t):
        if t % i == 0:
            print(False)
            break
    else:
        print(True)
```

```python
"""
输入两个正整数计算最大公因数和最小公倍数
"""

x = int(input('x: '))
y = int(input('y: '))
if x < y:
    m = x
else:
    m = y
for i in range(1,m+1):
    if x % i == 0 and y % i == 0:
        t = i
print(x * y / t)
```

```python
"""
找出100~999之间的所有水仙花数
水仙花数是各位立方和等于这个数本身的数
如: 153 = 1**3 + 5**3 + 3**3
"""

for i in range(100, 1000):
    a = i // 100
    b = i % 100 // 10
    c = i % 10
    if a*100 + b*10 + c == a**3 + b**3 + c**3:
        print(i)

for a in range(1,10):
    for b in range(0,10):
        for c in range(0,10):
            if a*100 + b*10 + c == a**3 + b**3 + c**3:
                print(a*100 + b*10 + c)
```
